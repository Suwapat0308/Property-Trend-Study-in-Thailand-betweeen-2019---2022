```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv(r'C\Python project\opendata_median_price.csv')
data.columns = ('Row','Transection_year','Transection_month','Provice_name_Eng','Provice_name_Th','District_name_Th','District_name_Eng',
'Property_type_Th','Property_type_Eng','Total','Median_price_sqm','Day')
data
```

```python
# Check null in dataframe
data.isna().sum().sum()
```

```python
data['Provice_name_Eng'] = data['Provice_name_Eng'].str.title()
data['District_name_Eng'] = data['District_name_Eng'].str.title()
data['Property_type_Eng'] = data['Property_type_Eng'].str.title()
data
```

```python
data['Transection_year_month'] = data['Transection_year'].astype(str) +"-"+ data['Transection_month'].astype(str)
data['Transection_year_month'] = pd.to_datetime(data['Transection_year_month'] , format='%Y-%m',errors='coerce').apply(lambda x: x.strftime('%Y-%m'))
data.sort_values(by='Transection_year_month',ascending=True,inplace=True)
data
```

```python
data['Transection_year_month'] = pd.to_datetime(data['Transection_year_month'] , format='%Y-%m',errors='coerce').apply(lambda x: x.strftime('%Y-%b'))
data
```

```python
# Check unique data
n = data.nunique(axis=0)
n
```

```python
Property_type = data.groupby('Property_type_Eng').count()
Property_type_count = Property_type['Row']
Property_type = Property_type_count.to_frame().reset_index(drop=False)

print ('Summary count of all property type as below \n',Property_type)
```

```python
Property_type_pv = data.groupby(['Provice_name_Eng','Property_type_Eng'], as_index=False).agg(count = ('Row' ,'count'))
Property_type_pv.reset_index(drop=False)
print ('Summary of each province and count property type as below \n',Property_type_pv)
```

```python
Property_type_pv
```

```python
# Data visualiztion
plt.figure(figsize=(15,8))
sns.set_style("whitegrid")


sns.barplot(x = 'Provice_name_Eng',       # x variable name
            y = 'count',       # y variable name
            hue = 'Property_type_Eng',  # group variable name
            data = Property_type_pv ,
            palette='tab10')     # dataframe to plot 

#add overall title
plt.title('Summary of each province and count property type', fontsize=16)
plt.legend(loc='upper right', title='Property_type')

#add axis titles
plt.xlabel('Count of property')
plt.ylabel('Province')
```

```python
data.sort_values(by='Transection_year_month', inplace=True)
Group_date = data.groupby('Transection_year_month').sum()
plt.figure(figsize=(15,8))
Group_date['Median_price_sqm'].plot()
plt.xlabel('Year Month Transsection')
plt.ylabel('Total Price')
plt.title('Overall Price', fontsize=16)
```

```python
# Data visualiztion
data.sort_values(by='Transection_year_month',ascending=True)
plt.figure(figsize=(15,8))

sns.set_style("whitegrid")

sns.lineplot(x = 'Transection_year_month',       # x variable name
            y = 'Total',       # y variable name
            hue = 'Property_type_Eng',  # group variable name
            data = data ,
            err_style=None ,sort = True, markers=True, style='Property_type_Eng', palette='husl')     # dataframe to plot 


#add overall title
plt.title('Time series plot', fontsize=16)
plt.legend(loc='upper right', title='Property_type')

#add axis titles
plt.xticks(rotation = 90)
plt.xlabel('Year-Month Transsection')
plt.ylabel('Total count')
```

```python
# Data visualiztion
data.sort_values(by='Transection_year_month',ascending=True)
plt.figure(figsize=(15,8))

sns.set_style("whitegrid")

sns.lineplot(x = 'Transection_year_month',       # x variable name
            y = 'Median_price_sqm',       # y variable name
            hue = 'Property_type_Eng',  # group variable name
            data = data ,
            err_style=None ,sort = True, markers=True, style='Property_type_Eng', palette='tab10')     # dataframe to plot 


#add overall title
plt.title('Time series plot of Property Price in Thailand (Baht)', fontsize=16)
plt.legend(loc='upper right', title='Property_type')

#add axis titles
plt.xticks(rotation = 90)
plt.xlabel('Year-Month Transsection')
plt.ylabel('price sqm')
```

```python
# Remove outlier to see bahavior
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

data = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
data
```
